```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df.drop('diagnosis',axis=1),df['diagnosis'],test_size=0.3,random_state=35)
```

```python
x_train.shape
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x_train_transformed=sc.fit_transform(x_train)
x_test_transformed=sc.transform(x_test)
```

```python
x_train_transformed
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(x_train_transformed,y_train)
```

```python
y_pred=lr.predict(x_test_transformed)
y_proba=lr.predict_proba(x_test_transformed)[:, 1]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, roc_auc_score, roc_curve,f1_score
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)
fpr, tpr, _ = roc_curve(y_test, y_proba)
```

```python
print(precision)
print(recall)
print(roc_auc)
print(fpr)
```

```python
plt.plot(fpr, tpr)  #roc curve
```

checking threshold value manually

```python
best_thresh = 0
best_f1 = 0
for t in np.arange(0.1, 0.9, 0.01):
    y_pred = (y_proba >= t).astype(int)
    score = f1_score(y_test, y_pred)
    if score > best_f1:
        best_f1 = score
        best_thresh = t


print(f"Best Threshold: {best_thresh:} with F1 Score: {best_f1:}")
y_pred_best = (y_proba >= best_thresh).astype(int)
print(f"Precision: {precision_score(y_test, y_pred_best):}")
print(f"Recall:    {recall_score(y_test, y_pred_best):}")
```
